```python
import matplotlib.pyplot as plt
```

```python
import datetime as dt
import matplotlib.patches as mpatches
import numpy as np

plt.figure(figsize=(10,5))


dates = []
for year in range(2016, 2017):
    for month in range(11, 13):
        dates.append(dt.datetime(year=year, month=month, day=1))
for year in range(2017, 2018):
    for month in range(1, 8):
        dates.append(dt.datetime(year=year, month=month, day=1))

red_patch = mpatches.Patch(color='green', label='Positive opinion')
green_patch = mpatches.Patch(color='red', label='Negative opinion')
plt.legend(handles=[red_patch,green_patch])
plt.xlabel('Month')
plt.ylabel('Sentiment Score')
plt.title('Positive vs Negative vs Neutral sentiments')
plt.plot(dates,pos_arr,'g')
plt.plot(dates,neu_arr,'y')
plt.plot(dates,neg_arr,'r')

print "POSITIVE = ",sum(pos_arr)/len(pos_arr),"NEGATIVE = ",sum(neg_arr)/len(pos_arr),"NEUTRAL = ",sum(neu_arr)/len(pos_arr)


plt.show()
```

```python
n_groups = 9
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.25
opacity = 0.8
 
rects1 = plt.bar(index, pos_arr, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Positive')
 
rects2 = plt.bar(index + bar_width, neg_arr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Negative')
rects3 = plt.bar(index + bar_width+bar_width, neu_arr, bar_width,
                 alpha=opacity,
                 color='y',
                 label='Neutral')

plt.xlabel('Month')
plt.ylabel('Sentiment Score')
plt.title('Positive vs Negative vs Neutral sentiments')
plt.xticks(index + bar_width, ('Nov-16', 'Dec-16', 'Jan-17', 'Feb-17','Mar-17','April-17','May-17','June-17','July-17'))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
 
plt.tight_layout()
plt.show()
```
